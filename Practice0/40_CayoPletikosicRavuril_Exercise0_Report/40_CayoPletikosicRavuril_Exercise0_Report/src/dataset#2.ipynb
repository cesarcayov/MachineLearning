## 5 Plots

````python
import matplotlib.pyplot as plt

plt.hist(y, bins=[-0.5, 0.5, 1.5], rwidth=0.8)
plt.xticks([0, 1])
plt.xlabel('Target')
plt.ylabel('Count')
plt.title('Target Histogram')

# Optional: show ratios in console
print(y.value_counts(normalize=True))
plt.show()
````

````python
# Combine X and y into one DataFrame (if not already)
df = X.copy()
df['target'] = y

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation
corr_with_target = numeric_df.corr()['target'].sort_values(ascending=False)
print(corr_with_target.head(6)) # top 5


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap (Numeric Features Only)')
plt.show()
````

````python
top_features = ['URLSimilarityIndex', 'HasSocialNet', 'HasCopyrightInfo', 'HasDescription', 'IsHTTPS', 'target']
plt.figure(figsize=(8,6))
sns.heatmap(df[top_features].corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Top 5 Correlated Features')
plt.show()
````

````python
top_corr = corr_with_target.head(5)

top_corr.plot(kind='barh', figsize=(6,4))
plt.title('Top 5 Correlated Features with Target')
plt.xlabel('Correlation')
plt.gca().invert_yaxis()
plt.show()
````

````python
plt.figure(figsize=(10,8))
top_features = ['HasSocialNet', 'HasCopyrightInfo', 'HasDescription', 'IsHTTPS']

for i, feat in enumerate(top_features):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=feat, hue='target', data=df)
    plt.title(f'{feat} by Target')

plt.tight_layout()
plt.show()
````

HasSocialNet: Phishing URLs (orange) overwhelmingly dominate when the feature is present.
HasCopyrightInfo: Phishing URLs strongly dominate at value 1.
HasDescription: Phishing URLs clearly dominate at value 1.
IsHTTPS: Phishing dominates at 1, but legit URLs appear more at 0 compared to other features.
