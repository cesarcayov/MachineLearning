```python
# =====================
# part 1: Import the dataset and required libraries
# =====================
import pandas as pd
from ucimlrepo import fetch_ucirepo
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
monk_s_problems = fetch_ucirepo(id=70) 
  
# data (as pandas dataframes) 
X_MONK = monk_s_problems.data.features 
Y_MONK = monk_s_problems.data.targets 
  
# metadata 
print(monk_s_problems.metadata) 
  
# variable information 
print(monk_s_problems.variables) 
```

```python
# =====================
# part 2: Inspect values
# =====================
X_MONK.info()
Y_MONK.info()
X_MONK.head()
Y_MONK.head()
```

```python
# =====================
# part 3: Check missing values
# =====================
# No missing values expected in this specific dataset (small)
print(X_MONK.isnull().sum())
print(Y_MONK.isnull().sum())
```

```python
# 1. the column ID can be deleted, because is an unique identifier
if 'ID' in X_MONK.columns:
    X_MONK = X_MONK.drop(columns=['ID'])

# 2. Convert target to Series (for scikit-learn)
y_MONK = Y_MONK.iloc[:, 0]

# 3. Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_MONK)

# 4. Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_MONK, test_size=0.2, random_state=42, stratify=y_MONK
)

# 5. Verify final preprocessed data
print("Clases y cantidad:\n", y_MONK.value_counts())
print("Proporción de clases:\n", y_MONK.value_counts(normalize=True))
```

```python
# Histogram of target classes using y_train
plt.figure(figsize=(6,4))
sns.countplot(x=y_train, hue=y_train, palette='pastel')
plt.title("Histogram of Target Classes (y_train)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
# Convertir a DataFrame si X_train es numpy
X_train_df = pd.DataFrame(X_train, columns=X_MONK.columns)

# Histogram of column 'a1'
plt.figure(figsize=(6,4))
sns.histplot(X_train_df['a1'], bins=10, kde=False)  # bins: número de barras
plt.title("Histogram of Feature a1")
plt.xlabel("a1 values")
plt.ylabel("Count")
plt.show()
```

```python
# === Export Cleaned Dataset (safe) ===
X = X_MONK.copy()
y = y_MONK.copy()

X['target'] = y
X.to_csv('dataset3_clean.csv', index=False)
print("✅ Saved cleaned dataset3_clean.csv safely")
```
